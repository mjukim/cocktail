```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup

# 웹 드라이버 초기화
driver = webdriver.Chrome()  # 크롬 드라이버를 사용하고 있다고 가정합니다.
driver.maximize_window()

# 페이지 열기
url = "https://iba-world.com/category/iba-cocktails/contemporary-classics/"
driver.get(url)

# 팝업 창 처리
try:
    popup = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[8]/p[2]/button[1]')))
    popup.click()
except:
    pass

# 스크롤을 하여 모든 컨텐츠 로드
SCROLL_PAUSE_TIME = 2

# 필요한 링크 추출
cocktail_links = driver.find_elements(By.CSS_SELECTOR, 'div.column.size-1of4 article[id^="post-"] a')

# 중복 방지를 위한 URL 추적
visited_urls = set()

# 각 링크에 접근하여 페이지 들어가고 뒤로가기로 나오기
cocktail_data2 = []

for index, link in enumerate(cocktail_links):
    cocktail_url = link.get_attribute("href")
    
    if cocktail_url in visited_urls:
        continue
    
    visited_urls.add(cocktail_url)
    
    # 링크를 열어 페이지에 들어가기
    driver.get(cocktail_url)
    print(f"Entered ({index+1}/{len(cocktail_links)}):", cocktail_url)
    
    # 페이지 로딩을 위해 시간 대기
    time.sleep(5)
    
    # 페이지 소스 가져오기
    page_source = driver.page_source
    
    # BeautifulSoup를 사용하여 파싱
    soup = BeautifulSoup(page_source, "html.parser")
    
    # 필요한 데이터 추출
    title = soup.find("div", class_="et_pb_title_container").get_text(strip=True)
    content_div = soup.find("div", class_="et_pb_module et_pb_post_content et_pb_post_content_0_tb_body blog-post-content")
    
    # 섹션 추출
    sections = content_div.find_all("h3")
    section_data2 = {}
    for section in sections:
        section_title = section.get_text(strip=True)
        section_content = section.find_next("p").get_text(strip=True)
        section_data2[section_title] = section_content
    
    # 데이터 저장
    cocktail_data2.append({
        "title": title,
        "sections": section_data2
    })
    
    # 페이지에서 뒤로 가기
    driver.execute_script("window.history.go(-1);")
    print(f"Exited ({index+1}/{len(cocktail_links)}):", cocktail_url)
    print("-----------------------------")
    
    time.sleep(5)
# 웹 드라이버 종료
driver.quit()

# 가져온 데이터 확인
for data in cocktail_data2:
    print("Title:", data["title"])
    print("Sections:")
    for section_title, section_content in data["sections"].items():
        print(f"- {section_title}: {section_content}")
    print("-----------------------------")
```

```python
with open("./cocktail_data2.json", "w", encoding="utf-8") as json_file:
    json.dump(cocktail_data2, json_file, ensure_ascii=False, indent=4)
```
